```python
df = pd.read_csv(r'./rus-covid-data.csv').set_index('date')
df.head()
```

```python
def plot_f(dataframe, features, rows=5, plotsize=(3, 3), sharex=False):
    reduced_df = dataframe[features]
    
    cols = len(features) // rows
    if len(features) % rows > 0:
        cols += 1
    
    if len(features) // rows == 0:
        rows, cols = len(features), 1

    print(rows, cols)
    y_size, x_size = plotsize[0]*rows + 5, plotsize[1]*cols + 5
    axes = reduced_df.plot(subplots=True, 
                           layout=(rows, cols),
                           figsize=(y_size, x_size),
                           sharex=sharex,
                           rot=90)
```

```python
plot_f(df, FULL_FEATURE_NAMES, plotsize=(5, 5), sharex=False)
```
